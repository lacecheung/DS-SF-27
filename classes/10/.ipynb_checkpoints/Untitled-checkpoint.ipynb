```python
import os

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
pd.set_option('display.notebook_repr_html', True)

import statsmodels.formula.api as smf
import seaborn as sns

from sklearn import linear_model, cross_validation

%matplotlib inline
```

```python
df = pd.read_csv(os.path.join('datasets', 'abalone.csv'))
```

```python
X = df.drop("Sex", axis = 1)
```

```python
X
```

```python
y = df.Sex
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape[0]


```

```python
X_test.shape[0]
```

```python

y_train.shape[0]
```

```python
y_test.shape[0]
```

```python
X_train.corr()

#We see that these dimensions are highly correlated, so we are choosing to use just length, diameter and rings as our model features
```

```python
color_wheel = {"I": "red",
               "M": "blue",
               "F": "green"}

color = y_train.map(color_wheel)

pd.tools.plotting.scatter_matrix(X_train[['Length', 'Height', 'Rings']], c=color, s=70, figsize = (10,10))
```

```python
features = X_train[["Height", "Length", "Rings", "WholeWeight"]]
c = y_train

model = linear_model.LogisticRegression().\
    fit(features, c)
```

```python
model.score(features, c)
```

```python
c_hat = model.predict(features)
```

```python
pd.crosstab(c_hat,
    c,
    rownames = ['Hypothesized Class'],
    colnames = ['True Class'])
```

```python
# We were able to identify the majority of infants, but the model could not separate M from Fs. 
# This makes sense from what we can see in the plots
```

```python
#lets try it with all the features added back

features_2 = X_train
c = y_train

model = linear_model.LogisticRegression().\
    fit(features_2, c)
    
model.score(features_2, c)
```

```python
c_hat_2 = model.predict(features_2)

pd.crosstab(c_hat_2,
    c,
    rownames = ['Hypothesized Class'],
    colnames = ['True Class'])
```

```python
# Better performance than before!
```

```python
# Are we including too many colinear variables? let's try it with just Height and Rings. 

features_3 = X_train[["Length", "Rings"]]
c = y_train

model = linear_model.LogisticRegression().\
    fit(features_3, c)
    
model.score(features_3, c)
```
